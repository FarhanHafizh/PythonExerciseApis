### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
users = pd.read_table('https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user',sep='|',index_col = 'user_id')
users.head()
```

```python
users.info()
```

### Step 4. Discover what is the mean age per occupation

```python
occ = users.groupby('occupation')
```

```python
occ['age'].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
users['gender'].dtype
```

```python
users[users['gender']=='M']['occupation'].count()
```

```python
users[users['gender']=='F']['occupation'].count()
```

```python
M_F_Ratio = (users[users['gender']=='M']['occupation'].count())/(users[users['gender']=='F']['occupation'].count())
print(M_F_Ratio)
```

```python
if users[users['gender']=='M']:
    return users['M'] == 1
else:
    return users['M'] == 0
```

```python
def gender(x):
    if x=='M':
        return 1
    else:
        return 0

users['M'] = users['gender'].apply(gender)
```

```python
occ_2 = users.groupby('occupation')
```

```python
(occ_2['M'].sum()/occ_2['M'].count()).sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
occ_2['age'].agg(['min','max'])
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users[(users['occupation']=='administrator') & (users['gender']=='F')]['age'].mean()
```

```python
occ_3 = users.pivot_table(values='age',index=['occupation','gender'])
occ_3
```

```python
users.groupby(['occupation','gender'])['age'].mean()
```
